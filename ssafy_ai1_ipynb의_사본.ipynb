````python
# ==============================
# 질문 유형 분석
# (pandas, matplotlib만 사용 - 기존 의존성 활용)
# ==============================

import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# 한글 폰트 (이미 설정했다면 생략)
plt.rc('font', family='NanumGothic')
plt.rcParams['axes.unicode_minus'] = False

# ==============================
# 1. 질문 유형 분류 함수
# ==============================
def classify_question_v3(q):
    q = str(q)
    if any(kw in q for kw in ["무엇", "뭐", "어떤 것", "어느 것"]):
        return "무엇 (객체 인식)"
    elif any(kw in q for kw in ["어디", "장소", "위치", "곳"]):
        return "어디 (장소/위치)"
    elif any(kw in q for kw in ["몇", "얼마", "수", "개"]):
        return "몇 개 (개수/수량)"
    elif any(kw in q for kw in ["누구", "사람", "인물"]):
        return "누구 (인물)"
    elif any(kw in q for kw in ["왜", "이유", "때문"]):
        return "왜 (이유)"
    elif any(kw in q for kw in ["어떻게", "방법", "방식"]):
        return "어떻게 (방법)"
    elif any(kw in q for kw in ["언제", "시간", "시기", "계절"]):
        return "언제 (시간)"
    elif any(kw in q for kw in ["무슨 색", "색깔", "색상"]):
        return "색깔"
    elif any(kw in q for kw in ["맞", "옳", "올바른", "아닌"]):
        return "참/거짓 판단"
    elif any(kw in q for kw in ["어떤", "어느", "종류", "형태", "유형"]):
        return "어떤 (속성/종류)"
    elif any(kw in q for kw in ["어떠한", "어떠"]):  # ✅ 추가
        return "어떤 (속성/종류)"                    # 같은 유형으로 통합
    else:
        return "기타"


# ==============================
# 2. 유형 분류 적용
# ==============================
train_df["question_type"] = train_df["question"].apply(classify_question_v3)

type_counts = train_df["question_type"].value_counts()
print("📊 질문 유형 분포:")
print(type_counts.to_string())
print(f"\n총 유형 수: {len(type_counts)}가지")

# ==============================
# 3. 유형별 분포 시각화
# ==============================
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# 막대 그래프
colors = plt.cm.Set2.colors
bars = axes[0].barh(type_counts.index, type_counts.values, color=colors[:len(type_counts)])
axes[0].set_title("질문 유형 분포", fontsize=13)
axes[0].set_xlabel("개수")
for bar in bars:
    w = bar.get_width()
    axes[0].text(w + 5, bar.get_y() + bar.get_height() / 2,
                 f"{int(w)}개 ({w/len(train_df)*100:.1f}%)",
                 va="center", fontsize=9)
axes[0].set_xlim(0, type_counts.max() * 1.3)

# 파이 차트
axes[1].pie(type_counts.values, labels=type_counts.index,
            autopct="%1.1f%%", colors=colors[:len(type_counts)],
            startangle=140)
axes[1].set_title("질문 유형 비율", fontsize=13)

plt.tight_layout()
plt.show()

# ==============================
# 4. 유형별 샘플 질문 출력
# ==============================
print("\n📝 유형별 샘플 질문 (각 2개):")
print("=" * 60)
for qtype in type_counts.index:
    samples = train_df[train_df["question_type"] == qtype]["question"].head(2).tolist()
    print(f"\n🔹 [{qtype}]")
    for s in samples:
        print(f"   - {s}")

# ==============================
# 5. 유형별 정답 분포 (편향 확인)
# ==============================
print("\n\n📊 유형별 정답 선지 분포:")
type_answer = train_df.groupby(["question_type", "answer"]).size().unstack(fill_value=0)
display(type_answer)

fig, ax = plt.subplots(figsize=(12, 5))
type_answer.plot(kind="bar", ax=ax, color=["#4C9BE8","#E87C4C","#4CE87C","#E84C9B"])
ax.set_title("유형별 정답 선지 분포", fontsize=13)
ax.set_xlabel("질문 유형")
ax.set_ylabel("개수")
ax.legend(title="정답 선지")
plt.xticks(rotation=30, ha="right")
plt.tight_layout()
plt.show()
````

````python
# ==============================
# 1. 추가 패키지 설치
# (기존 의존성에 없는 것만)
# ==============================
!pip -q install qwen-vl-utils

# ==============================
# 2. 임포트
# ==============================
import torch
import pandas as pd
from PIL import Image
from transformers import Qwen2_5_VLForConditionalGeneration, AutoProcessor
# ⚠️ Qwen3-VL은 transformers 4.57.0 기준
# 현재 git 버전으로 설치했으므로 아래 import로 시도
try:
    from transformers import Qwen3VLForConditionalGeneration
    MODEL_CLASS = Qwen3VLForConditionalGeneration
    print("✅ Qwen3VL 클래스 로드 성공")
except ImportError:
    from transformers import Qwen2_5_VLForConditionalGeneration
    MODEL_CLASS = Qwen2_5_VLForConditionalGeneration
    print("⚠️ Qwen3VL 미지원 → Qwen2.5VL 클래스로 대체")

from qwen_vl_utils import process_vision_info

# ==============================
# 3. 모델 로드 (4bit 양자화)
# ==============================
MODEL_NAME = "Qwen/Qwen3-VL-2B-Instruct"

from transformers import BitsAndBytesConfig

bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_compute_dtype=torch.float16,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4"
)

print(f"🔄 모델 로딩 중: {MODEL_NAME}")
model = MODEL_CLASS.from_pretrained(
    MODEL_NAME,
    quantization_config=bnb_config,
    device_map="auto"
)
processor = AutoProcessor.from_pretrained(MODEL_NAME)
model.eval()
print("✅ 모델 로드 완료!")

# ==============================
# 4. 추론 함수
# ==============================
IMAGE_DIR = "/content/"   # path 컬럼이 "train/train_0001.jpg" 형식이므로 루트만 지정

def predict(row):
    img_path = IMAGE_DIR + row["path"]

    # 프롬프트 구성
    prompt = f"""다음 이미지를 보고 질문에 답하세요.
질문: {row['question']}
선택지:
a. {row['a']}
b. {row['b']}
c. {row['c']}
d. {row['d']}

반드시 a, b, c, d 중 하나만 답하세요. 다른 말은 하지 마세요."""

    messages = [
        {
            "role": "user",
            "content": [
                {"type": "image", "image": img_path},
                {"type": "text",  "text": prompt}
            ]
        }
    ]

    # 입력 준비
    text = processor.apply_chat_template(
        messages, tokenize=False, add_generation_prompt=True
    )
    image_inputs, video_inputs = process_vision_info(messages)
    inputs = processor(
        text=[text],
        images=image_inputs,
        videos=video_inputs,
        return_tensors="pt"
    ).to(model.device)

    # 추론
    with torch.no_grad():
        generated_ids = model.generate(
            **inputs,
            max_new_tokens=5,    # a/b/c/d 한 글자만 필요
            do_sample=False
        )

    generated_ids_trimmed = [
        out[len(inp):] for inp, out in zip(inputs.input_ids, generated_ids)
    ]
    output = processor.batch_decode(
        generated_ids_trimmed,
        skip_special_tokens=True,
        clean_up_tokenization_spaces=False
    )[0].strip().lower()

    # a/b/c/d 추출
    for ch in ["a", "b", "c", "d"]:
        if ch in output:
            return ch
    return "a"  # 추출 실패 시 기본값

# ==============================
# 5. 샘플 테스트 (3개)
# ==============================
print("\n🧪 샘플 추론 테스트:")
for i in range(3):
    row = train_df.iloc[i]
    pred = predict(row)
    correct = "✅" if pred == row["answer"] else "❌"
    print(f"{correct} [{i+1}] 예측: {pred} | 정답: {row['answer']} | 질문: {row['question'][:30]}...")

# ==============================
# 6. 전체 Test 추론 & CSV 제출 파일 생성
# ==============================
print("\n🚀 전체 추론 시작 (test 3887개)...")
from tqdm import tqdm

results = []
for _, row in tqdm(test_df.iterrows(), total=len(test_df)):
    pred = predict(row)
    results.append({"id": row["id"], "answer": pred})

submission = pd.DataFrame(results)
submission.to_csv("/content/submission.csv", index=False)
print(f"\n🎉 제출 파일 저장 완료! → /content/submission.csv")
print(submission.head())
print(f"\n정답 분포:\n{submission['answer'].value_counts().sort_index()}")
````

````python
import os, re, math, random
import pandas as pd
from PIL import Image
from torch.utils.data import Dataset, DataLoader
from dataclasses import dataclass
import torch
from typing import Dict, List, Any
from transformers import (
    Qwen2_5_VLForConditionalGeneration,
    AutoProcessor,
    BitsAndBytesConfig,
    get_linear_schedule_with_warmup
)
from peft import LoraConfig, get_peft_model, prepare_model_for_kbit_training
from tqdm import tqdm

# 이미지 로드 시 픽셀 제한 해제
Image.MAX_IMAGE_PIXELS = None

# 디바이스 GPU 우선 사용 설정
device = "cuda" if torch.cuda.is_available() else "cpu"
print("Device:", device)

# 사전 학습 모델 정의
MODEL_ID = "Qwen/Qwen2.5-VL-3B-Instruct"
IMAGE_SIZE = 384
MAX_NEW_TOKENS = 8
SEED = 42
random.seed(SEED); torch.manual_seed(SEED); torch.cuda.manual_seed_all(SEED)

# 데이터셋 로드
base_path = "/content/drive/MyDrive/ssafy-ai/"
train_df = pd.read_csv(os.path.join(base_path, "train.csv"))
test_df  = pd.read_csv(os.path.join(base_path, "test.csv"))

# 경로를 절대 경로로 수정 (중요: FileNotFoundError 해결)
train_df['path'] = train_df['path'].apply(lambda x: os.path.join(base_path, x))
test_df['path'] = test_df['path'].apply(lambda x: os.path.join(base_path, x))

# 학습데이터 200개만 추출
train_df = train_df.sample(n=200, random_state=SEED).reset_index(drop=True)

print("Data Loaded and Paths updated.")
display(train_df.head(2))
````

````python
# 모델 지시사항
SYSTEM_INSTRUCT = (
    "You are a helpful visual question answering assistant. "
    "Answer using exactly one letter among a, b, c, or d. No explanation."
)

# 프롬프트
def build_mc_prompt(question, a, b, c, d):
    return (
        f"{question}\n"
        f"(a) {a}\n(b) {b}\n(c) {c}\n(d) {d}\n\n"
        "Please output the answer using only one lowercase letter from a, b, c, or d."
    )
````

````python
# 커스텀 데이터셋
class VQAMCDataset(Dataset):
    def __init__(self, df, processor, train=True):
        self.df = df.reset_index(drop=True)
        self.processor = processor
        self.train = train

    def __len__(self): return len(self.df)

    def __getitem__(self, i):
        row = self.df.iloc[i]
        img = Image.open(row["path"]).convert("RGB")

        q = str(row["question"])
        a, b, c, d = str(row["a"]), str(row["b"]), str(row["c"]), str(row["d"])
        user_text = build_mc_prompt(q, a, b, c, d)

        messages = [
            {"role":"system","content":[{"type":"text","text":SYSTEM_INSTRUCT}]},
            {"role":"user","content":[
                {"type":"image","image":img},
                {"type":"text","text":user_text}
            ]}
        ]
        if self.train:
            gold = str(row["answer"]).strip().lower()
            messages.append({"role":"assistant","content":[{"type":"text","text":gold}]})

        return {"messages": messages, "image": img}

# 데이터 콜레이터
@dataclass
class DataCollator:
    processor: Any
    train: bool = True

    def __call__(self, batch):
        texts, images = [], []
        for sample in batch:
            messages = sample["messages"]
            img = sample["image"]

            text = self.processor.apply_chat_template(
                messages,
                tokenize=False,
                add_generation_prompt=False
            )
            texts.append(text)
            images.append(img)

        enc = self.processor(
            text=texts,
            images=images,
            padding=True,
            return_tensors="pt"
        )

        if self.train:
            enc["labels"] = enc["input_ids"].clone()

        return enc
````

````python
# 검증용 데이터 분리
split = int(len(train_df)*0.9)
train_subset, valid_subset = train_df.iloc[:split], train_df.iloc[split:]

# VQAMCDataset 형태로 변환
train_ds = VQAMCDataset(train_subset, processor, train=True)
valid_ds = VQAMCDataset(valid_subset, processor, train=True)

# 데이터로더
train_loader = DataLoader(train_ds, batch_size=1, shuffle=True, collate_fn=DataCollator(processor, True), num_workers=0)
valid_loader = DataLoader(valid_ds, batch_size=1, shuffle=False, collate_fn=DataCollator(processor, True), num_workers=0)
````

````python
from tqdm.auto import tqdm

model = model.to(device)
GRAD_ACCUM = 4

# 옵티마이저, 학습 스케줄러
optimizer = torch.optim.AdamW(model.parameters(), lr=1e-4)
num_training_steps = 1 * math.ceil(len(train_loader)/GRAD_ACCUM)
scheduler = get_linear_schedule_with_warmup(optimizer, int(num_training_steps*0.03), num_training_steps)

# 스케일러
scaler = torch.cuda.amp.GradScaler(enabled=True)

# 학습 루프
global_step = 0
for epoch in range(1):
    running = 0.0
    progress_bar = tqdm(train_loader, desc=f"Epoch {epoch+1} [train]", unit="batch")
    for step, batch in enumerate(progress_bar, start=1):
        batch = {k:v.to(device) for k,v in batch.items()}
        with torch.cuda.amp.autocast(dtype=torch.bfloat16):
            outputs = model(**batch)
            loss = outputs.loss / GRAD_ACCUM

        scaler.scale(loss).backward()
        running += loss.item()

        if step % GRAD_ACCUM == 0:
            scaler.step(optimizer)
            scaler.update()
            optimizer.zero_grad(set_to_none=True)
            scheduler.step()
            global_step += 1

            avg_loss = running / GRAD_ACCUM
            progress_bar.set_postfix({"loss": f"{avg_loss:.3f}"})
            running = 0.0

    model.eval()
    val_loss = 0.0
    val_steps = 0
    with torch.no_grad(), torch.cuda.amp.autocast(dtype=torch.bfloat16):
        for vb in tqdm(valid_loader, desc=f"Epoch {epoch+1} [valid]", unit="batch"):
            vb = {k:v.to(device) for k,v in vb.items()}
            val_loss += model(**vb).loss.item()
            val_steps += 1
    print(f"[Epoch {epoch+1}] valid loss {val_loss/val_steps:.4f}")
    model.train()

# 모델 저장
SAVE_DIR = "/content/drive/MyDrive/ssafy-ai/qwen2_5_vl_3b_lora"
model.save_pretrained(SAVE_DIR)
processor.save_pretrained(SAVE_DIR)
print("Saved:", SAVE_DIR)
````
